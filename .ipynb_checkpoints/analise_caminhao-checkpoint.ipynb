```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
import matplotlib.pyplot as plt
import datetime
```

```python
file = "logs_escavadeira_18_08_21.csv"
df = pd.read_csv(f"dados\{file}", encoding='utf-8')
df.columns = ["latitude", "longitude", "status", "angle", "fuel", "weight", "velocity", "imei", "date_created"]
df["date_created"] = pd.to_datetime(df["date_created"])
```

# An√°lise da Escavadeira Dia 18/08

```python
pd.set_option('display.max_rows', df.shape[0]+1)
#pd.set_option('display.max_rows', 60)
df.sort_values(by="date_created", ascending=True, inplace=True)
df_obra = df[(df["date_created"] > "2021-08-17 00:00:00") & (df["latitude"] < -1)]
dia_obra = df_obra["date_created"]
diff_time = dia_obra.diff()
df_obra.insert(9, "atraso", diff_time, True)
df_obra
```

## Correla√ß√£o entre √Çngulo e N√≠vel de Combust√≠vel

```python
df_corr = df_obra.loc[:,["angle","fuel","atraso"]]
df_corr.corr()
```

## Gr√°fico de Atraso entre Amostras

```python
diff_time_s = df_obra["atraso"] / np.timedelta64(1, 's') #convert to seconds
diff_time_list = diff_time_s.to_list()
#print(f"Atrasos: {diff_time_list[1:]}")
print(f"Atraso M√°ximo: {diff_time.max()} \
      \nAtraso M√≠nimo: {diff_time.min()} \
      \nAtraso M√©dio: {diff_time.mean()}")

plt.close("all")
ts = pd.Series(diff_time[1:], index=list(range(1,len(diff_time))))
ts.plot()
```
